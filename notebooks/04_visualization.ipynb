```python
# 04_visualization.ipynb
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load cleaned data safely
clean_csv_path = 'data/processed/trending_videos_clean.csv'
if not os.path.exists(clean_csv_path):
    raise FileNotFoundError(f"{clean_csv_path} not found. Please run 02_data_cleaning.ipynb first.")

df = pd.read_csv(clean_csv_path)
```

```python
# Top 10 videos by views
top_views = df.sort_values('views', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='views', y='title', data=top_views, palette='viridis')
plt.title('Top 10 YouTube Trending Videos by Views')
plt.xlabel('Views')
plt.ylabel('Video Title')
plt.show()
```

```python
# Top 10 videos by engagement ratio
top_engagement = df.sort_values('engagement_ratio', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='engagement_ratio', y='title', data=top_engagement, palette='magma')
plt.title('Top 10 YouTube Trending Videos by Engagement Ratio')
plt.xlabel('Engagement Ratio')
plt.ylabel('Video Title')
plt.show()
```

```python
# Scatter plot: Views vs Likes
plt.figure(figsize=(8,6))
sns.scatterplot(x='views', y='likes', data=df)
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# Scatter plot: Title Length vs Engagement Ratio
plt.figure(figsize=(8,6))
sns.scatterplot(x='title_length', y='engagement_ratio', data=df)
plt.title('Title Length vs Engagement Ratio')
plt.xlabel('Title Length')
plt.ylabel('Engagement Ratio')
plt.show()
```
